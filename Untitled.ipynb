````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,101)
y = np.sin(x)
````

````python
plt.plot(x,y)
plt.grid()
plt.title('Sinus')
````

````python
y = np.cos(x)
plt.plot(x,y)
plt.grid()
plt.title('Cosinus')
````
